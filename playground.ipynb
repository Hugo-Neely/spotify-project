```python
import os
import spotipy
from spotipy.oauth2 import SpotifyOAuth
import numpy as np
import pandas as pd
```

```python
scope = "playlist-private-read"

sp = spotipy.Spotify(auth_manager=SpotifyOAuth(scope=scope, 
                                               client_id=os.getenv('SPOTIPY_CLIENT_ID'),
                                               client_secret=os.getenv('SPOTIPY_CLIENT_SECRET'),
                                               redirect_uri=os.getenv('SPOTIPY_REDIRECT_URI')))

results = sp.current_user_saved_tracks()
for idx, item in enumerate(results['items']):
    track = item['track']
    print(idx, track['artists'][0]['name'], " â€“ ", track['name'])
```

```python
scope = "playlist-read-private"

sp = spotipy.Spotify(auth_manager=SpotifyOAuth(scope=scope, 
                                               client_id=os.getenv('SPOTIPY_CLIENT_ID'),
                                               client_secret=os.getenv('SPOTIPY_CLIENT_SECRET'),
                                               redirect_uri=os.getenv('SPOTIPY_REDIRECT_URI')))
```

```python
playlists_0_49 = sp.current_user_playlists(limit = 50, offset = 0)
playlists_50_99 = sp.current_user_playlists(limit = 50, offset = 50)
playlists_100_149 = sp.current_user_playlists(limit = 50, offset = 100)
playlists_150_199 = sp.current_user_playlists(limit = 50, offset = 150)  # up to 163
```

```python
df_pl = pd.concat([pd.DataFrame(playlists_0_49['items']), 
                   pd.DataFrame(playlists_50_99['items']), 
                   pd.DataFrame(playlists_100_149['items']), 
                   pd.DataFrame(playlists_150_199['items'])])

# indexes for each year
ind_18 = df_pl['name'].str.contains('2018')
ind_19 = df_pl['name'].str.contains('2019')
ind_20 = df_pl['name'].str.contains('2020')
ind_21 = df_pl['name'].str.contains('2021')
ind_22 = df_pl['name'].str.contains(' 22')
ind_23 = df_pl['name'].str.contains(' 23')
ind_24 = df_pl['name'].str.contains('-24')
ind_25 = df_pl['name'].str.contains('2025')

# all monthly playlists
df_monthly = df_pl[ind_18 | ind_19 | ind_20 | ind_21 | ind_22 | ind_23 | ind_24 | ind_25].copy()
df_monthly
```
